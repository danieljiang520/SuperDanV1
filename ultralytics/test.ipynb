```python
from ultralytics import YOLO
from PIL import Image
import cv2
import torch
```

```python
model = YOLO('./runs/detect/train12/weights/best.pt') # select a .pt file in runs/detect/train #num/weights/best.pt
results = model.predict(source="https://www.youtube.com/watch?v=2y0u0BUvc3M", stream=True, conf=0.1) # save in runs/detect/track #num
# ttps://www.youtube.com/watch?v=2y0u0BUvc3M
```

```python
boxes = []
print_now = True
cnt=0
for r in results:
    cnt += 1
    box = r.boxes
    conf = box.conf
    xyxy = box.xyxy
    if len(conf) == 0:
        continue
    conf_max_xyxy = xyxy[torch.argmax(conf)]

    # center of the bounding box (x, y)
    center = torch.tensor([(conf_max_xyxy[0] + conf_max_xyxy[2]) / 2, (conf_max_xyxy[1] + conf_max_xyxy[3])  / 2])
    boxes.append(center)
    '''
    if print_now: 
        print(f"conf_max_xyxy: {conf_max_xyxy}")
        print(f"center: {center}")
        print_now = False
        '''
print(cnt)
```

```python
print(cnt)
```

```python
boxes = torch.stack(boxes, dim=0)
```

```python
# 1 is going up, 0 is going down

net_height = torch.mean(boxes[:, 1])
print(net_height)

count = 0
# -1 is down, 1 is up area
curr_side = -1 if boxes[0][1] < net_height else 1
for box in boxes:
    if box[1] > net_height:
        side = 1
    else:
        side = 0

    if side != curr_side:
        count += 1
        curr_side = side
print(count)



# STEP = 3
# for i in range(0, len(boxes) - STEP, STEP):
#     dx = boxes[i + STEP][0] - boxes[i][0]
#     dy = boxes[i + STEP][1] - boxes[i][1]
#     print(dx, dy)
# print(boxes.shape)
    







```
